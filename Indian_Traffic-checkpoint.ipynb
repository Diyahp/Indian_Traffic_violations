```python
# Data Shape and Structure
print("Data Shape:", df.shape)
print("\nData Types:\n", df.dtypes)

# Missing Values
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100
print("\nMissing Values:\n", missing_values)
print("\nMissing Value Percentage:\n", missing_percentage)

# Descriptive Statistics
print("\nDescriptive Statistics:\n", df.describe(include='all'))

# Data Distribution of Categorical Variables
categorical_cols = ['Violation_Type', 'Location', 'Vehicle_Type', 'Vehicle_Color', 'Registration_State', 'Driver_Gender', 'Weather_Condition', 'Road_Condition', 'Issuing_Agency', 'License_Validity', 'Helmet_Worn', 'Seatbelt_Worn', 'Traffic_Light_Status', 'Breathalyzer_Result', 'Towed', 'Fine_Paid', 'Payment_Method', 'Court_Appearance_Required']
for col in categorical_cols:
    print(f"\nDistribution of {col}:\n{df[col].value_counts()}")

# Visualize Top 10 Violation Types
import matplotlib.pyplot as plt
violation_counts = df['Violation_Type'].value_counts().head(10)
plt.figure(figsize=(10, 6))
violation_counts.plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Frequent Violation Types')
plt.xlabel('Violation Type')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Examine Date and Time Distribution
print("\nDate Range:", df['Date'].min(), "to", df['Date'].max())
print("\nTime Distribution:\n", df['Time'].value_counts().head(10)) #Example
```

```python
#Data cleaning.
#clean the data by handling missing values and standardizing text fields.
#Handling missing values, standardlize text flields, and convert date/time columns to the correct data types.
#the addresses the first three instructions of thhe subtask.
```

```python
import pandas as pd

# Handle Missing Values
# Impute missing values in 'Helmet_Worn' and 'Seatbelt_Worn' with the mode
for col in ['Helmet_Worn', 'Seatbelt_Worn']:
    df[col] = df[col].fillna(df[col].mode()[0])

# Remove rows with missing values in other columns (if any)
df.dropna(inplace=True)


# Standardize Text Fields
text_cols = ['Violation_Type', 'Location', 'Vehicle_Type', 'Vehicle_Color',
             'Registration_State', 'Driver_Gender', 'Weather_Condition',
             'Road_Condition', 'Issuing_Agency', 'License_Validity',
             'Helmet_Worn', 'Seatbelt_Worn', 'Traffic_Light_Status',
             'Breathalyzer_Result', 'Towed', 'Fine_Paid', 'Payment_Method',
             'Court_Appearance_Required']

for col in text_cols:
    df[col] = df[col].str.strip().str.title()  # Remove extra whitespace and title-case


# Data Type Conversion
try:
    df['Date'] = pd.to_datetime(df['Date'])
    df['Time'] = pd.to_datetime(df['Time'], format='%H:%M').dt.time
except ValueError as e:
    print(f"Error converting date/time: {e}")

# Remove future dates
df = df[df['Date'].dt.year <= 2025]

display(df.head())
```

```python
#The step is to detect and the handle outliers in numerical columns like "Fine_Amount", and "Driver_Age","Recorded_speed" 
```

```python
import numpy as np

# Outlier Detection and Handling
numerical_cols = ['Fine_Amount', 'Driver_Age', 'Recorded_Speed']

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]

display(df.head())
```

```python
#Data analysis
#Analyze the cleaned traffic violation data to answer specific questions outlined in the subtask.
#Analyze the cleaned traffic violation data to answer the specific questions outlined in the subtask. This involves identifying the most frequent violation types, investigating location and time patterns, exploring correlations between violation type and time, and analyzing geographic distribution. 
#I will use value_counts(), groupby(), and plotting functions to achieve this.
 
```

```python
# 1. Most Common Violations
top_violations = df['Violation_Type'].value_counts().head(5)
print("Top 5 Most Common Violations:\n", top_violations)

# 2. Location and Time Patterns (Illustrative example - Drunk Driving)
import matplotlib.pyplot as plt
drunk_driving_violations = df[df['Violation_Type'] == 'Drunk Driving']

location_counts = drunk_driving_violations['Location'].value_counts()
plt.figure(figsize=(10, 5))
location_counts.plot(kind='bar', color='skyblue')
plt.title('Locations of Drunk Driving Violations')
plt.xlabel('Location')
plt.ylabel('Number of Violations')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


time_counts = drunk_driving_violations['Time'].value_counts().head(10)
plt.figure(figsize=(10, 5))
time_counts.plot(kind='bar', color='orange')
plt.title('Time of Day for Drunk Driving Violations (Top 10)')
plt.xlabel('Time')
plt.ylabel('Number of Violations')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# 3. Violation Type and Time Correlation (Illustrative example)
# Group data by violation type and hour of the day
df['Hour'] = df['Time'].apply(lambda x: x.hour)
violation_time_correlation = df.groupby(['Violation_Type', 'Hour'])['Violation_ID'].count().unstack()

# Plot the correlation
violation_time_correlation.plot(kind='bar', figsize=(15, 8))
plt.title('Violation Type and Time of Day Correlation')
plt.xlabel('Violation Type')
plt.ylabel('Number of Violations')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Hour of Day')
plt.tight_layout()
plt.show()


# 4. Geographic Distribution
location_violations = df['Location'].value_counts()
print("\nGeographic Distribution (Top 10):\n", location_violations.head(10))

plt.figure(figsize=(12, 6))
location_violations.head(10).plot(kind='bar', color='green')
plt.title('Top 10 Locations with Highest Violation Counts')
plt.xlabel('Location')
plt.ylabel('Number of Violations')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
